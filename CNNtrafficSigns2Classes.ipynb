```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def grid_display(list_of_images, list_of_titles=[], no_of_columns=2, figsize=(10,10), cmap='None'):
    fig = plt.figure(figsize=figsize)
    column = 0
    for i in range(len(list_of_images)):
        column += 1
        if column == no_of_columns+1:
            fig = plt.figure(figsize=figsize)
            column = 1
        fig.add_subplot(1, no_of_columns, column)
        plt.imshow(list_of_images[i], cmap)
        plt.axis('off')
        if len(list_of_titles) >= len(list_of_images):
            plt.title(list_of_titles[i])
```

```python
input_image1 = cv2.imread('train/child_1.png') 
input_image1 = cv2.cvtColor(input_image1, cv2.COLOR_BGR2RGB)
input_image2 = cv2.imread('train/brick_1.png') 
input_image2 = cv2.cvtColor(input_image2, cv2.COLOR_BGR2RGB)
input_image3 = cv2.imread('train/child_131.png') 
input_image3 = cv2.cvtColor(input_image3, cv2.COLOR_BGR2RGB)
input_image4 = cv2.imread('train/brick_131.png') 
input_image4 = cv2.cvtColor(input_image4, cv2.COLOR_BGR2RGB)
input_image5 = cv2.imread('train/child_594.png') 
input_image5 = cv2.cvtColor(input_image5, cv2.COLOR_BGR2RGB)
input_image6 = cv2.imread('train/brick_594.png') 
input_image6 = cv2.cvtColor(input_image6, cv2.COLOR_BGR2RGB)
titles = ["input_image1", 
          "input_image2",
         "input_image3",
         "input_image4",
         "input_image5",
          "input_image6"] 
images = [input_image1, 
          input_image2,
         input_image3,
         input_image4,
         input_image5,
         input_image6]
print(type(input_image1))

grid_display(images, titles, 3, (6,6), 'gray')
```

```python
train_child = ['train/{}'.format(i) for i in os.listdir('train') if 'child' in i][100:]
train_brick = ['train/{}'.format(i) for i in os.listdir('train') if 'brick' in i][100:]
test_img = ['train/{}'.format(i) for i in os.listdir('train')][:100]
train_img = train_brick[:] + train_child[:]

print(" - : --- " + str(len(train_child))) 
print(" -  : --- " + str(len(train_brick))) 
print(" : --- " + str(len(test_img))) 
print(" : --- " + str(len(train_img))) 
```

```python
def read_process(list_of_img):
    x = [] # imgs
    y = [] # labels
    for img in list_of_img:
        x.append(cv2.resize(cv2.imread(img, cv2.IMREAD_COLOR), (48,48), interpolation=cv2.INTER_CUBIC))
        if 'child' in img:
            y.append(1)
        elif 'brick' in img:
            y.append(0)
    return x,y
x, y = read_process(train_img)
x[0]
y
```

```python
plt.figure(figsize=(20,20))
columns = 5
for i in range(columns):
    plt.subplot(5 / columns+1, columns, i+1)
    plt.imshow(x[i])
    
```

```python
x = np.array(x)
y = np.array(y)

print('x: ', x.shape, 'y: ', y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.20, random_state = 2)
print('x_train '+ str(x_train.shape))
print('x_valid '+ str(x_valid.shape))
print('y_train'+ str(y_train.shape))
print('y_valid'+ str(y_valid.shape))
```

```python
len_train = len(x_train)
len_val = len(x_valid)
print(len_train, len_val)
batch_size = 3 
```

```python
# Importing the Keras libraries and packages
from keras.models import Sequential #   
from keras.layers import Conv2D #   
from keras.layers import MaxPooling2D #       
from keras.layers import Flatten #       , 
#           
from keras.layers import Dense #   
from keras.layers import BatchNormalization
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape = (48, 48, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(BatchNormalization())

model.add(Conv2D(64, (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2))) 
model.add(BatchNormalization())

model.add(Conv2D(128, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2)))
model.add(BatchNormalization())

model.add(Conv2D(128, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2)))
model.add(BatchNormalization())

model.add(Flatten())

model.add(Dense(units = 512, activation = 'relu')) 
model.add(Dense(units = 1, activation = 'sigmoid')) 

# Optimizer parameter is to choose the stochastic gradient descent algorithm 
# (  -     )

# Loss parameter is to choose the loss function
#   -   

# the metrics parameter is to choose the performance metric
#   -    
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array, load_img


train_datagen = ImageDataGenerator(rescale = 1./255,
shear_range = 0.2,
zoom_range = 0.2,
horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

train_gener = train_datagen.flow(x_train, y_train, batch_size=batch_size)
var_gener = test_datagen.flow(x_valid, y_valid, batch_size=batch_size)
```

```python
history = model.fit_generator(train_gener, steps_per_epoch=len_train // batch_size, epochs=25, validation_data=var_gener, validation_steps=len_val // batch_size)
```

```python
model.save_weights('model_wieghts25FINALep.h5')
model.save('model_keras25FINALep.h5')
```

```python
acc = history.history['acc']
valid_acc = history.history['val_acc']
loss = history.history['loss']
valid_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.figure()
plt.plot(epochs, loss, 'green', label='Training loss')
plt.plot(epochs, valid_loss, 'orange', label='Validation loss')
plt.title('train and valid loss')
plt.legend()

plt.show()
```

```python
plt.figure()
plt.plot(epochs, acc, 'green', label='Training acc')
plt.plot(epochs, valid_acc, 'orange', label='Validation acc')
plt.title('train and valid acc')
plt.legend()

plt.show()
```
